```csharp

using System.Text.Json;
using System.Text.Json.Serialization;
using System;
using System.Collections.Generic;
using System.Linq;
using System.Text;
using System.Threading.Tasks;
using OfficeOpenXml;
using System.IO;
using Microsoft.DotNet.Interactive.Formatting;
using System.Collections.Generic;
using System.Text.RegularExpressions;
Formatter.ListExpansionLimit=2000;
//Excel文件所在的地址
FileInfo file = new FileInfo(@"C:\Users\hiyan\Downloads\c_machine_position.xlsx");
ExcelPackage.License.SetNonCommercialPersonal("My Name");//个人
using (ExcelPackage excelPackage = new ExcelPackage(file))
{
    //指定需要读入的sheet名
    ExcelWorksheet excelWorksheet = excelPackage.Workbook.Worksheets["c_machine_position"];
    //比如读取第一行,第一列的值数据
    object a = excelWorksheet.Cells[1, 1].Value;
    //读取第一行,第二列的值为
    object b = excelWorksheet.Cells[1, 3].Value;
    //然后根据需要对a，b转为字符串，或者double，int等..
    var rows = excelWorksheet.Dimension.End.Row;
    List<string> line = [];
    for(int i = 1;i<=rows ;i++){
        var id = excelWorksheet.Cells[i, 1].Value;
        var name =excelWorksheet.Cells[i, 3].Value.ToString();
        var type =excelWorksheet.Cells[i, 2].Value;
        if((!type.ToString().Equals("1"))&&(!type.ToString().Equals("7")))
        {
            var nums = name.Split("-");
            var m = "";
            if("1".Equals(nums[0]))
            {
                m += $"tc1_preposition_{(Convert.ToInt16(nums[2])-1)/4+1}_ad_all";
                if("1".Equals(nums[1]))
                {
                    if(Convert.ToInt16(nums[2])%4 == 1)
                    {
                        m=m+"_6";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 2)
                    {
                        m=m+"_7";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 3)
                    {
                        m=m+"_8";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 0)
                    {
                        m=m+"_14";
                    }
                }
                if("2".Equals(nums[1]))
                {
                    if(Convert.ToInt16(nums[2])%4 == 1)
                    {
                        m=m+"_1";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 2)
                    {
                        m=m+"_2";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 3)
                    {
                        m=m+"_3";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 0)
                    {
                        m=m+"_4";
                    }
                }
                if("3".Equals(nums[1]))
                {
                    if(Convert.ToInt16(nums[2])%4 == 1)
                    {
                        m=m+"_9";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 2)
                    {
                        m=m+"_10";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 3)
                    {
                        m=m+"_11";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 4)
                    {
                        m=m+"_12";
                    }
                }
            }
            if("2".Equals(nums[0]))
            {
                m += $"mp1_preposition_{(Convert.ToInt16(nums[2])-1)/4+1}_ad_all";
                                if("1".Equals(nums[1]))
                {
                    if(Convert.ToInt16(nums[2])%4 == 1)
                    {
                        m=m+"_6";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 2)
                    {
                        m=m+"_7";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 3)
                    {
                        m=m+"_8";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 0)
                    {
                        m=m+"_14";
                    }
                }
                if("2".Equals(nums[1]))
                {
                    if(Convert.ToInt16(nums[2])%4 == 1)
                    {
                        m=m+"_1";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 2)
                    {
                        m=m+"_2";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 3)
                    {
                        m=m+"_3";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 0)
                    {
                        m=m+"_4";
                    }
                }
                if("3".Equals(nums[1]))
                {
                    if(Convert.ToInt16(nums[2])%4 == 1)
                    {
                        m=m+"_9";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 2)
                    {
                        m=m+"_10";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 3)
                    {
                        m=m+"_11";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 0)
                    {
                        m=m+"_12";
                    }
                }
            }
            if("3".Equals(nums[0]))
            {
                m += $"m1_preposition_{(Convert.ToInt16(nums[2])-1)/4+1}_ad_all";
                                if("1".Equals(nums[1]))
                {
                    if(Convert.ToInt16(nums[2])%4 == 1)
                    {
                        m=m+"_6";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 2)
                    {
                        m=m+"_7";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 3)
                    {
                        m=m+"_8";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 0)
                    {
                        m=m+"_14";
                    }
                }
                if("2".Equals(nums[1]))
                {
                    if(Convert.ToInt16(nums[2])%4 == 1)
                    {
                        m=m+"_1";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 2)
                    {
                        m=m+"_2";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 3)
                    {
                        m=m+"_3";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 0)
                    {
                        m=m+"_4";
                    }
                }
                if("3".Equals(nums[1]))
                {
                    if(Convert.ToInt16(nums[2])%4 == 1)
                    {
                        m=m+"_9";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 2)
                    {
                        m=m+"_10";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 3)
                    {
                        m=m+"_11";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 0)
                    {
                        m=m+"_12";
                    }
                }
            }
            if("4".Equals(nums[0]))
            {
                m += $"m2_preposition_{(Convert.ToInt16(nums[2])-1)/4+1}_ad_all";
                                if("1".Equals(nums[1]))
                {
                    if(Convert.ToInt16(nums[2])%4 == 1)
                    {
                        m=m+"_6";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 2)
                    {
                        m=m+"_7";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 3)
                    {
                        m=m+"_8";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 0)
                    {
                        m=m+"_14";
                    }
                }
                if("2".Equals(nums[1]))
                {
                    if(Convert.ToInt16(nums[2])%4 == 1)
                    {
                        m=m+"_1";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 2)
                    {
                        m=m+"_2";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 3)
                    {
                        m=m+"_3";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 0)
                    {
                        m=m+"_4";
                    }
                }
                if("3".Equals(nums[1]))
                {
                    if(Convert.ToInt16(nums[2])%4 == 1)
                    {
                        m=m+"_9";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 2)
                    {
                        m=m+"_10";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 3)
                    {
                        m=m+"_11";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 0)
                    {
                        m=m+"_12";
                    }
                }
            }
            if("5".Equals(nums[0]))
            {
                m += $"mp2_preposition_{(Convert.ToInt16(nums[2])-1)/4+1}_ad_all";
                                if("1".Equals(nums[1]))
                {
                    if(Convert.ToInt16(nums[2])%4 == 1)
                    {
                        m=m+"_6";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 2)
                    {
                        m=m+"_7";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 3)
                    {
                        m=m+"_8";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 0)
                    {
                        m=m+"_14";
                    }
                }
                if("2".Equals(nums[1]))
                {
                    if(Convert.ToInt16(nums[2])%4 == 1)
                    {
                        m=m+"_1";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 2)
                    {
                        m=m+"_2";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 3)
                    {
                        m=m+"_3";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 0)
                    {
                        m=m+"_4";
                    }
                }
                if("3".Equals(nums[1]))
                {
                    if(Convert.ToInt16(nums[2])%4 == 1)
                    {
                        m=m+"_9";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 2)
                    {
                        m=m+"_10";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 3)
                    {
                        m=m+"_11";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 0)
                    {
                        m=m+"_12";
                    }
                }
            }
            if("6".Equals(nums[0]))
            {
                m += $"tc2_preposition_{(Convert.ToInt16(nums[2])-1)/4+1}_ad_all";
                                if("1".Equals(nums[1]))
                {
                    if(Convert.ToInt16(nums[2])%4 == 1)
                    {
                        m=m+"_6";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 2)
                    {
                        m=m+"_7";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 3)
                    {
                        m=m+"_8";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 0)
                    {
                        m=m+"_14";
                    }
                }
                if("2".Equals(nums[1]))
                {
                    if(Convert.ToInt16(nums[2])%4 == 1)
                    {
                        m=m+"_1";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 2)
                    {
                        m=m+"_2";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 3)
                    {
                        m=m+"_3";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 0)
                    {
                        m=m+"_4";
                    }
                }
                if("3".Equals(nums[1]))
                {
                    if(Convert.ToInt16(nums[2])%4 == 1)
                    {
                        m=m+"_9";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 2)
                    {
                        m=m+"_10";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 3)
                    {
                        m=m+"_11";
                    }
                    if(Convert.ToInt16(nums[2])%4 == 0)
                    {
                        m=m+"_12";
                    }
                }
            }
            m+=$":{excelWorksheet.Cells[i, 4].Value}_{excelWorksheet.Cells[i, 2].Value}_{excelWorksheet.Cells[i, 1].Value}";
            line.Add(m);
        }
        
    }
    
    Console.WriteLine(JsonSerializer.Serialize(ConvertToGroupDictionary(line)));
    //Console.WriteLine(JsonSerializer.Serialize(line));
}

    /// <summary>
    /// 将字符串数组转换为分组字典
    /// </summary>
    /// <param name="dataArray">输入数据数组</param>
    /// <returns>分组字典</returns>
    public Dictionary<string, Dictionary<string, string>> ConvertToGroupDictionary(List<string> dataArray)
    {
        var result = new Dictionary<string, Dictionary<string, string>>();
        var regex = new Regex(@"^(?<group>.+?)_(?<key>\d+):(?<value>.+)$", RegexOptions.Compiled);

        foreach (var item in dataArray)
        {
            var match = regex.Match(item);
            if (!match.Success) continue;

            var groupName = match.Groups["group"].Value;
            var key = match.Groups["key"].Value;
            var value = match.Groups["value"].Value;

            // 确保分组存在
            if (!result.ContainsKey(groupName))
            {
                result[groupName] = new Dictionary<string, string>();
            }

            // 添加键值对（重复键会自动覆盖）
            result[groupName][key] = value;
        }

        return result;
    }
```
